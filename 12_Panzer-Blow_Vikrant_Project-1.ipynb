# 0. Introduction

In this project, we are going to use a non-linear weighted least squares optimization approach to solve the problem of getting a better estimate of our robot's trajectory. Least squares formulations are widely used for optimization, be it computer vision or robotics or machine learning. We will dive deep into it during this project and you will have complete clarity on optimization for vector-valued residual functions. 

In this "Introduction" section, I am going to provide an introduction for SLAM problem for a robot operating in the 2D world. It is 2. section in this Project. The 1D SLAM problem (1.) is far much simpler to understand and will be described directly in the 1. section. 

In a 2D world, a robot has 3 degrees of freedom, i.e. its pose in the world can be expressed by the state vector $\mathbf{x}=(x, y, \theta)^{\mathrm{T}}$. For the scope of this project, we are interested only in the robot's trajectory through the $2 \mathrm{D}$ world, and NOT in distinct landmarks or the surronding map of the environment, i.e. we are only interested in "L"ocalization part of SLAM. 

Therefore, we can represent it as a graph where the vertices represent robot poses $\mathbf{x}_{i}$ and edges represent the spatial constraints between these poses. Such a map is generally called a pose graph.

Two different kinds of constraints are necessary for pose graph SLAM. The first are
odometric constraints that connect two successive states $\mathbf{x}_{i}$ and $\mathbf{x}_{i+1}$ via a motion model. Furthermore, in order to perform loop closing, the robot has to recognize places it already visited before. This place recognition is also a part of the front-end and provides the second type of constraint, the loop closure constraints. These constraints connect two not necessarily successive poses $\mathbf{x}_{i}$ and $\mathbf{x}_{j}$.

```python
def fx(x,th,dx,dy):
    dx=jnp.reshape(dx,(-1,1))
    dy=jnp.reshape(dy,(-1,1))
    return x+dx*jnp.cos(th)-dy*jnp.sin(th) 

def fy(y,th,dx,dy):
    dx=jnp.reshape(dx,(-1,1))
    dy=jnp.reshape(dy,(-1,1))
    return y+dy*jnp.cos(th)+dx*jnp.sin(th) 

def fth(th,dt):
    dt=jnp.reshape(dt,(-1,1))
    return th+dt

# def get_f(X,Y,THETA,edges):
def get_f(xx,edges):
    X,Y,THETA=jnp.reshape(xx,(3,len(xx)//3))
    X=jnp.reshape(X,(-1,1))
    Y=jnp.reshape(Y,(-1,1))
    THETA=jnp.reshape(THETA,(-1,1))
    X=jnp.array(X)
    Y=jnp.array(Y)
    THETA=jnp.array(THETA)
    i=edges[0]
    j=edges[1]
    dx=edges[2]
    dy=edges[3]
    dt=edges[4]
    ocl=len(X)
    #odometry constrain
    ocx=fx(X[:-1],THETA[:-1],dx[:ocl-1],dy[:ocl-1])-X[1:]
    ocy=fy(Y[:-1],THETA[:-1],dx[:ocl-1],dy[:ocl-1])-Y[1:]
    octh=fth(THETA[:-1],dt[:ocl-1])-THETA[1:]
    #loop constrain
    ocl-=1
#     print(i[ocl:])
#     iloop=i[ocl:].reshape(-1, 1)
#     jloop=j[ocl:].reshape(1, -1)
    iloop=i[ocl:]
    jloop=j[ocl:]
    lcx=fx(X[iloop],THETA[iloop],dx[ocl:],dy[ocl:])-X[jloop]
    lcy=fy(Y[iloop],THETA[iloop],dx[ocl:],dy[ocl:])-Y[jloop]
    lcth=fth(THETA[iloop],dt[jloop])-THETA[jloop]
    #initial constrains
    icx=X[0]-(-5)
    icy=Y[0]-(-8)
    icth=THETA[0]-(0)
    icx=jnp.reshape(icx,(-1,1))
    icy=jnp.reshape(icy,(-1,1))
    icth=jnp.reshape(icth,(-1,1))
    return jnp.concatenate((ocx,ocy,octh,lcx,lcy,lcth,icx,icy,icth)).reshape(-1)

def getOmega(odo=100.0, loop=500,init=1000 ,odoedges=119, loopedges=20):
    omega = jnp.zeros((odoedges*3 + loopedges*3+3, odoedges*3 + loopedges*3+3))
    for i in range(odoedges*3 + loopedges*3 + 3):
        if i < odoedges*3:
            omega = jax.ops.index_update(omega, jax.ops.index[i, i], odo)
        elif i < odoedges*3 + loopedges*3 :
            omega = jax.ops.index_update(omega, jax.ops.index[i, i], loop)
        else :
            omega = jax.ops.index_update(omega, jax.ops.index[i, i], init)
            
    return omega
    
    
def analitical_jac(THETA,edges):
    i=edges[0]
    j=edges[1]
    dx=edges[2]
    dy=edges[3]
    dt=edges[4]
    ocl=len(THETA)
    iloop=i[ocl-1:]
    jloop=j[ocl-1:]
    jaco=[]
    #X
    for x in range(0,ocl):
        dfx=jnp.zeros(420).tolist()
        if x==0:
            dfx[0]=1
            dfx[-3]=1
        elif x==ocl-1:
            dfx[118]=-1
        else:
            dfx[x]=1
            dfx[x-1]=-1
        for i in range(0,20):
            if iloop[i]==x:
                dfx[357+i]=1
        for j in range(0,20):
            if jloop[j]==x:
                dfx[357+j]=-1
        
        jaco.append(dfx)
        
    #Y
    for y in range(0,ocl):
        dfy=jnp.zeros(420).tolist()
        if y==0:
            dfy[119+0]=1
            dfy[-2]=1
        elif y==ocl-1:
            dfy[237]=-1
        else:
            dfy[119+y]=1
            dfy[119+y-1]=-1
        
        for i in range(0,20):
            if iloop[i]==y:
                dfy[357+20+i]=1
        for j in range(0,20):
            if jloop[j]==y:
                dfy[357+20+j]=-1
        jaco.append(dfy)
        
    #THETA
    for t in range(0,ocl):
        dft=jnp.zeros(420).tolist()
        if t<119:
            dft[t] = (-dx[t]*jnp.sin(THETA[t])-dy[t]*jnp.cos(THETA[t])).tolist()[0]
            dft[119+t] = (-dy[t]*jnp.sin(THETA[t])+dx[t]*jnp.cos(THETA[t])).tolist()[0]
        if t==0:
            dft[238+t]=1
            dft[-1]=1
        elif t==119:
            dft[237+119]=-1
        else:
            dft[238+t]=1
            dft[237+t]=-1
            
        for i in range(0,20):
            if iloop[i]==t:
                dft[357+i] = (-dx[119+i]*jnp.sin(THETA[t])-dy[119+i]*jnp.cos(THETA[t])).tolist()[0]
                dft[357+20+i] = (-dy[119+i]*jnp.sin(THETA[t])+dx[119+i]*jnp.cos(THETA[t])).tolist()[0]
                dft[357+40+i] = 1
        
        for j in range(0,20):
            if jloop[j]==t:
                dft[357+40+j] = -1
        
        jaco.append(dft)
    
    return jnp.array(jaco).T
        
                        
def frobNorm(P1, P2, str1="mat1", str2="mat2"):
    jnp.set_printoptions(suppress=True)
    val = jnp.linalg.norm(P1 - P2, 'fro')
    print(f"Frobenius norm between {str1} and {str2} is: {val}")        
    
        
def lm(X,Y,THETA,edges,w,lamda=100,max_itr=100,tol=0.001):
    cost_itr=[]
    x_itr=[]
    for it in range(0,max_itr):
#         af=get_f(X,Y,THETA,edges) 
#         ff = lambda X,Y,THETA: get_f(X,Y,THETA,edges)
#         J = jacfwd(ff)
#         j = J(X,Y,THETA)
        #(j'wj + lamda.I)dx = -j'w'f
#         print(X[edges[]])
        xx=jnp.concatenate((X,Y,THETA))
        xx=xx.reshape(-1)
        f=get_f(xx,edges)
        ff = lambda aa : get_f(aa,edges)
        J=jacfwd(ff)
        j=J(xx)
#         w=getOmega()
        jj=analitical_jac(THETA,edges)
        frobNorm(j,jj)
        H = j.T@w@j + lamda*jnp.eye(360)
        b = j.T@w.T@f
        dx = -jnp.linalg.inv(H)@b
        dX,dY,dT = jnp.reshape(dx,(3,len(dx)//3))
        X+=dX.reshape((-1,1))
        Y+=dY.reshape((-1,1))
        THETA+=dT.reshape((-1,1))
        cost_itr.append(0.5*f.T@w@f)
        print("err",it+1,"=",cost_itr[-1])
        x_itr.append([X,Y,THETA])
#         draw(X,Y,THETA)
        if it!=0:
            if cost_itr[-1]>cost_itr[-2]:
                lamda/=10
            else:
                lamda*=10
            if abs(cost_itr[-1]-cost_itr[-2])<tol:
                break
    return cost_itr,x_itr

w=getOmega(1,10,100)
cost,x=lm(jnp.array(X),jnp.array(Y),jnp.array(THETA),e,w,lamda=0.1) 
print("YAAAAA")
```

1) SLAM addresses the problem of simultaneous estimation of state of robot along with mapping the environment desribing it's features, landmarks, obstacles etc. There might be some information rendered to us before hand, and some, which we percieve through sensors like IMU, GPS, Lidar, LOC, Map, RGB, Stereo and so on. The approach to SLAM can be thought and disected into ages  - _classic age(1986-2004)_  which introduced the probabilistic formulation of SLAM, EKF, MLE etc. The _algorithmic-analysis(2004-2015)_ age studied the key properties of SLAM like observability,convergence,  and  consistency. In the modern _robust-perception age_ is characterized by robust performance, high level understanding, resource awareness and task-driven perception. 

The current formulation of SLAM, uses MAP along with factor-graphs, and a resemblance to Bundle Adjustment problem(BA). A SLAM system might be fragile in  algorithmic or  hardware-related aspect.  The  former  class includes  failure  modes  induced  by  limitation  of  the  existing SLAM  algorithms  (e.g.,  difficulty  to  handle  extremely  dy-namic or harsh environments). The latter includes failures due to to sensor or actuator degradation. One of the main sources of algorithmic failures is data association. 

The following Open Problems were discussed in context of SLAM - 
- _Failsafe SLAM and recover_ : Virtually  all  SLAM  techniques  are  based  on  iterative optimization of nonconvex costs. Due to this,   outlier  rejection  outcome  depends  on  the  quality  ofthe initial guess and the system isinherently fragile. A reliable system is a system which is **fail-safe** and **fault-aware**. The current SLAM algorithms do not that guarantee this.
- _Robustness to HW failure_ : The SLAM system must be tolerant to sensor errors and conflicts. We need to detect  degraded  sensor  operationa andadjust sensor noise statistics (covariances, biases) accordingly. We also need to resolve  conflicting  informationfrom  different  sensors.
- _Metric Relocalization_ : For estimating the relative pose with respect to the  previously  built  map, we still use feature-based  approaches. We need to add sufficient variance to work reliably, maybe using trajectory matching and mapping  with  one  sensor  modality.
- _Time  varying  and  deformable  maps_ :  A real-world environment, may have time-varying environment, and maps which might be rigid. We need to accomodate to such cases. An ideal SLAM solution should be able to reason about dynamics in the environment including non-rigidity, work overlong time periods generating “all terrain” maps and be able todo  so  in  real  time.
- _Automatic parameter tuning_ :If SLAM has to work “out of the box” in  arbitrary  scenarios,  methods  for  automatic  tuning  of  the involved parameters need to be considered. Currently, RANSAC, control feature matching like parameters make it sufficiently complex and which is why there is a need to have automation in this.

SLAM solvers must be _scalable_ , say it could be used to solve  ocean exploration for environmental monitoring, non-stop cleaning robots in our ever changing cities, or large-scaleprecision  agriculture. The following open problems arrive in factor graph optimization - 
- _Map representation_ : It is an important topic to understand how data is stored on the cloud, raw representations  as  point  clouds  or  volumetric  maps. 
- _Learning, forgetting, remembering_:  Robot needs to remember and store the memory of the map. Certain parts can be "swapped off" and some parts can be retained. 
- _Robust  distributed  mapping_ : In a Multi-SLAM case, we need to properly map multi-robots understanding.
- _Resource-constrained  platforms_ : Where to run the SLAM? In mobile or small computing platforms have RAM based constraints. When designing micro  aerial  vehicles,  or  robotic  insect, this comes into handy.

SLAM also deals with semantic formulation. We addess some of the formulation here - 

- _Consistent semantic-metric fusion_ :  We need to incorporate  the  confidence  or  uncertainty  of  thesemantic categorization in the best possible way.
- _Semantic  mapping  is  much  more  than  a  categorizationproblem_ : How  to  representthese properties, and interrelationships, are questions to answerfor high level human-robot interaction.
- _Ignorance,  awareness,  and  adaptation_: Given  some  priorknowledge,  the  robot  should  be  able  to  reason  about  newconcepts and their semantic representation.
- _Semantic-based reasoning_ : Robots must be able to reason the environmental placement similar to humans, this reasoning ability might provide them crucial context. 

In the context of deep learning, we have the following open problems - 
- _Perceptual  tool_: Semantic labeling of pixels in a variety of scene types reachesperformance  levels  of  around  80%  accuracy  or  more. Perception is assumed to be a black box in SLAM, but till how much.
- _Practical   deployment_ : To get good GPUs and computing.
- _Online and life-long learning_ : Deep  networks  are  usually  trained  on  closed-world scenarios with, say, a fixed number of object classes. A significant challenge is to harness the power of deep networks in  a  one-shot  or  zero-shot  scenario.
